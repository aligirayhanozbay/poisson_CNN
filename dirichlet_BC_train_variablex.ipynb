```python
import tensorflow as tf
from Dirichlet_BC_NN import *#Dirichlet_BC_NN, Dirichlet_BC_NN_2
from Lp_integral_norm import Lp_integral_norm
from generate_laplace_soln import generate_laplace_soln
import numpy as np
import matplotlib.pyplot as plt
tf.keras.backend.set_floatx('float64')
```

```python
#mod = Dirichlet_BC_NN()
mod = Dirichlet_BC_NN_2D(data_format = 'channels_first', mae_component_weight = 1e+0, n_quadpts=27)
mod((tf.random.uniform((10,1,74), dtype = tf.keras.backend.floatx()), tf.random.uniform((10,1), dtype = tf.keras.backend.floatx()), tf.constant(np.random.randint(48,96))))
from IPython.display import clear_output
clear_output()
mod.load_weights('Dirichlet_BC_NN_variablex4.h5')
#mod = Dirichlet_BC_NN_3()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-4)
```

```python
def dataset_generator():
    batch_size = 40
    while True:
        nx = np.random.randint(64,85)
        ny = np.random.randint(64,85)
#         print('------------------------')
#         print('generated data shape: ' + str(nx) + ' ' + str(ny) + '\n')
#         model.x_output_resolution = nx
#         print('x output res from generator: ' + str(model.x_output_resolution) + '\n')
        boundaries, soln, dx = generate_laplace_soln(batch_size=batch_size, nonzero_boundaries=['left'], smoothness=np.random.randint(5,20), nx = nx, ny = ny)
#         print('soln shape: ' + str(soln.shape))
        yield [boundaries['left'], dx * tf.ones((soln.shape[0],1), dtype = 'float64'), tf.constant(nx)], soln
```

```python
cb = [tf.keras.callbacks.ModelCheckpoint('Dirichlet_BC_NN_variablex4.h5', monitor='mse', verbose=1, save_best_only=True, save_weights_only=True), tf.keras.callbacks.ReduceLROnPlateau(monitor='loss', min_lr = 1e-7, verbose = True, patience = 5)]
mod.compile(loss = mod.integral_loss, optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-6), metrics = ['mse', 'mae'])
mod.run_eagerly = True
mod.fit_generator(generator=dataset_generator(), steps_per_epoch=200, epochs=5000, validation_data=dataset_generator(), validation_steps=3, callbacks=cb)
```

```python
i = dataset_generator()
inp, soln = next(i)
```
